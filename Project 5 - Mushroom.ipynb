```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
mushroom = pd.read_csv("D:/Jobin/DataTrained/Project/Mushroom.csv")
mushroom
```

Columns are not numbers. So we need to convert them

```python
mappings = list()
encoder = LabelEncoder()
for column in range(len(mushroom.columns)):
    mushroom[mushroom.columns[column]] = encoder.fit_transform(mushroom[mushroom.columns[column]])
    mappings_dict = {index: label for index, label in enumerate(encoder.classes_)}
    mappings.append(mappings_dict)
```

```python
Y = mushroom["Class"]
X = mushroom.drop('Class', axis = 1)
```

```python
Y.shape
```

```python
Y = Y.values.reshape(-1,1)
Y.shape
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X))
```

```python
X
```

Data for X is scaled to make them in similar range

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 150)
```

```python
LgR = LogisticRegression()
LgR.fit(X_train, Y_train)
```

```python
predictions = LgR.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(Y_test, predictions)
accuracy
```

```python
print(classification_report(Y_test, predictions))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print('Mean absolute error:', mean_absolute_error(Y_test, predictions))
print('Mean squared error:', mean_squared_error(Y_test, predictions))
print('Root Mean squared error:', np.sqrt(mean_squared_error(Y_test, predictions)))
print('R2 score is:', r2_score(Y_test, predictions))
```

```python
svc = SVC()
svc.fit(X_train, Y_train)
svc.score(X_train, Y_train)
predsvc = svc.predict(X_test)
print(accuracy_score(Y_test, predsvc))
print(confusion_matrix(Y_test, predsvc))
print(classification_report(Y_test, predsvc)) 
```
