Some columns are skewed

```python
from scipy.stats import zscore
z = np.abs(zscore(sonar))
z
```

```python
threshold = 3
print(np.where(z>3))
```

Above are the outliers

```python
sonar_new = sonar[(z<3).all(axis = 1)]
sonar_new
```

```python
sonar.shape
```

```python
sonar_new.shape
```

Removed outliers

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(sonar_new.drop("Class", axis = 1), sonar_new["Class"], test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LgR = LogisticRegression()
LgR.fit(X_train, Y_train)
```

```python
predictions = LgR.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
accuracy = confusion_matrix(Y_test, predictions)
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test, predictions)
accuracy
```

```python
print(classification_report(Y_test, predictions))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print('Mean absolute error:', mean_absolute_error(Y_test, predictions))
print('Mean squared error:', mean_squared_error(Y_test, predictions))
print('Root Mean squared error:', np.sqrt(mean_squared_error(Y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_test, predictions))
```
